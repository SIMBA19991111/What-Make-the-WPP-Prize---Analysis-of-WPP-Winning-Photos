```python
from transformers import pipeline
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
summ = pipeline("summarization",model = "philschmid/flan-t5-base-samsum",device=0)
```

```python
df = pd.read_csv("dfimageurlall.csv")
df = df[df['Index'] == 1]
df = df.reset_index(drop=True)
```

```python
df
```

```python
summary_list = []
for i in range (0,len(df)):
    try:
        summary = summ(df['Description'][i], min_length=5, do_sample=True)
        summary_list.append(str(summary[0]['summary_text']))
    except:
        try:
            summary = summ(df['Background'][i]+df['Story'][i], min_length=5, do_sample=True)
            summary_list.append(str(summary[0]['summary_text']))
        except:
            summary_list.append("")
            continue
```

```python
summary_df = pd.DataFrame(summary_list)
display(summary_df)
```

```python
for i in range (2815,len(df)):
        summary = summ(df['Background'][i]+df['Story'][i], min_length=5, do_sample=True)
        summary_list.append(str(summary[0]['summary_text']))
```

```python
with pd.option_context('display.max_colwidth', -1):
    print(summary_df)
```

```python
df['Summary'] = summary_list
df.to_csv("dfimageurlall_summerized.csv",index=False)
```
