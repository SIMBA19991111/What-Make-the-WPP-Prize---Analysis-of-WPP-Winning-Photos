```python
from geopy.geocoders import Nominatim
import pycountry

def get_country(location_name):
    geolocator = Nominatim(user_agent="my_app")
    location = geolocator.geocode(location_name, exactly_one=True, addressdetails=True)
    while location is not None and 'address' in location.raw and ('city' in location.raw['address'] or 'town' in location.raw['address']):
        if 'village' in location.raw['address']:
            return pycountry.countries.get(alpha_2=location.raw['address']['country_code']).name
        elif 'city' in location.raw['address']:
            return pycountry.countries.get(alpha_2=location.raw['address']['country_code']).name
        elif 'region' in location.raw['address']:
            return pycountry.countries.get(alpha_2=location.raw['address']['country_code']).name
        elif 'county' in location.raw['address']:
            return pycountry.countries.get(alpha_2=location.raw['address']['country_code']).name
        elif 'state' in location.raw['address']:
            return pycountry.countries.get(alpha_2=location.raw['address']['country_code']).name
        elif 'country' in location.raw['address']:
            return pycountry.countries.get(alpha_2=location.raw['address']['country_code']).name
        else:
            location = geolocator.geocode(location.raw['address'], exactly_one=True, addressdetails=True)
    else:
        return None
```

```python
location_list = []
location_probability_list = []
for i in range(0,len(df)):
    result = ie_en(df['Summary'][i])
    try:
        result_item =  [elem['text'] for item in result for elem in item['Location']]
        result_probability = [elem['probability'] for item in result for elem in item['Location']]
        for i in range(0,len(result_item)):
            location = []
            location_probability = []
            location.append(get_country(result_item[i]))
            location_probability.append(result_probability[i])
        location = str(list(set(location)))
        location_list.append(location)
        location_probability = str(list(set(location_probability)))
        location_probability_list.append(location_probability)
    except:
        location_list.append('None')
        location_probability_list.append('None')
```

```python
location = pd.DataFrame({'location':location_list,'location_probability':location_probability_list})
location['location_probability'] = location['location_probability'].apply(lambda x: x.replace('[',''))
location['location_probability'] = location['location_probability'].apply(lambda x: x.replace(']',''))
location['location_probability'] = location['location_probability'].astype(float,errors='ignore')
location['location']=location['location'].str.replace('[','')
location['location']=location['location'].str.replace(']','')
location['location']=location['location'].str.replace("'",'')
location.sample(n=50)
```

```python
df['Location'] = location['location']
df['Location_probability'] = location['location_probability']
df
```
