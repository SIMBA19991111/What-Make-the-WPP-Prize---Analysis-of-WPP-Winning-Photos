**Recap of awards for all countries**

```python
df_filtered3 = pd.read_csv("df_filtered_final_pro.csv", encoding="UTF-8")
```

```python
len(df_filtered3)
```

```python
df_filtered3.head(5)
```

```python
df_filtered3 = df_filtered3[df_filtered3['Photographer_Country'] != 'Unknown']
grouped = df_filtered3.groupby('Award Year')['Photographer_Country'].value_counts()
grouped = grouped.sort_values(ascending=False)

grouped
```

```python
grouped = df_filtered3.groupby('Award Year')['Photographer_Country'].value_counts()
grouped = grouped.sort_values(ascending=False)
grouped = grouped.reset_index(name='Counts')  
grouped = grouped.set_index(['Award Year', 'Photographer_Country'])
grouped.index.set_names(['Award Year', 'Photographer_Country'], inplace=True)  
```

```python
grouped
```

```python
import pandas as pd

# Using pivot_table
pivot = pd.pivot_table(grouped, values='Counts', index='Photographer_Country', columns='Award Year', aggfunc='count', fill_value=0)

# Calculate the cumulative quantity for each country
cumulative_sum = pivot.cumsum(axis=1)
```

```python
cumulative_sum
```

```python
countries = pd.read_csv('https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/master/all/all.csv')
countries = countries[['alpha-2', 'name']]
countries.columns = ['Photographer_Country', 'Country Name']
df = cumulative_sum

# Convert the indices of the matrix dataframe to country names
df = df.rename(index=countries.set_index('Photographer_Country')['Country Name'])

df
```

```python
df.to_csv('my_data.csv', index=True)
```

```python
df_filtered4 = pd.read_csv("Data_url.csv", encoding="UTF-8")
```

```python
df_filtered4
```

```python
df['Region'] = ''
df['Image URL'] = ''

for index, row in df.iterrows():
    country_name = df_filtered4[df_filtered4['Country Name'] == index]['Country Name']
    
    # If the corresponding Country Name is found, add the values ​​​​of the corresponding Region and Image URL columns to a new column of df
    if not country_name.empty:
        region = df_filtered4.loc[country_name.index[0], 'Region']
        image_url = df_filtered4.loc[country_name.index[0], 'Image URL']
        df.at[index, 'Region'] = region
        df.at[index, 'Image URL'] = image_url
    else:
        df.at[index, 'Region'] = 'Unknown'
        df.at[index, 'Image URL'] = 'Unknown'
```

```python
df
```

```python
df.to_csv('my_data2.csv', index=True)
```

**Photograhper**

```python
grahper_count = pd.read_csv("df_filtered_final_pro.csv", encoding="UTF-8")
```

```python
grahper_count = grahper_count.loc[:, ['Photographer', 'Photographer_Country']]
grahper_count
```

```python
# Group count for Photographer and Photographer_Country columns
grahper_count = grahper_count.groupby(['Photographer', 'Photographer_Country']).size().reset_index(name='Count')

# Arrange in reverse order according to the counting results of photographers, and view the top 20 most photographers
grahper_count = grahper_count.sort_values(by=['Count'], ascending=False).head(20)
```
